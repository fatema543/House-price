```python
df = pd.read_csv('/kaggle/input/house-price/Housing.csv')
df
```

```python
x_feature = ['area','bedrooms','bathrooms','stories']
x_train = np.array(df[x_feature])
y_train = np.array(df['price'])
```

```python
fig,ax=plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(x_train[:,i],y_train)
    ax[i].set_xlabel(x_feature[i])
ax[0].set_ylabel("Price (1000's)")
plt.show()
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_train,y_train)
pred = linear_reg.predict(x_train)
b = linear_reg.intercept_
w = linear_reg.coef_
print("Cost: ", pred - y_train)
```
